Q4-)1)Factorial of 13Using Recursion and without iteration


```python
def fact(num):
    fact=1
    if num == 0:
        return 1

    for i in range(1,num+1):        
        fact = fact*i
        
    return fact
```

```python
def factr(n):
      
    if n == 0:
        return 1
     
    return n * factr(n-1)
```

```python
fact(2)
```

```python
def fact(num):
    lst=[]
    fact=1
    if num == 0:
        return 1

    for i in range(1,int(num/4)):        
        fact = fact*i
        lst.append(fact)
    if(num in lst):
        print("Yes")
    else:
        print("No")
    return 
```
